```python
### 3) ---------------------------------------------------------------
# convert X_train from shape (60000,28,28) --> (60000, 784)
# convert X_test from shape (10000,28,28) --> (10000, 784)
# flatten each 28x28 to a single 784 vector

X_train_reshaped = X_train.reshape((X_train.shape[0], 784))
#X_train.shape[0] gets the # of training samples.
# reshape reorganizes the array without copying.
X_test_reshaped = X_test.reshape((X_test.shape[0], 784))

# chose not to do a for loop because its slow and has to iterate over 600000 values - I tried and it took too long or got stuck.

# confirming reshape
print("Original X_train shape:", X_train.shape)
print("Reshaped X_train shape:", X_train_reshaped.shape)
print("Original X_test shape:", X_test.shape)
print("Reshaped X_test shape:", X_test_reshaped.shape)
# Nx784 matrix where N=60000 for train or 10000 for test.

### 4) --------------------------------------------------------
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

k_values = [1,3,5,7,9,11,15]
accuracies = []

print("Testing different k values... \n")
for k in k_values:
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train_reshaped, y_train) # train the classifier on training data
  y_pred = knn.predict(X_test_reshaped) # making predictions -
  # accuracy of k values
  accuracy = accuracy_score(y_test, y_pred)
  accuracies.append(accuracy)
  print(f"k={k:2d}: Accuracy = {accuracy * 100:.2f}%")
```

The color photo could be represented as a 3D matrix of 3, 2D vectors stacked on each other where the third dimension contains RGB color channels ranging from 0 to 255. I would then flatten each image in order to reshape it to (n images, height * width * 3). The result is each row with one image with (height * width * 3) features. This creates a standard tabular format where each row is an observation and each column os a feature (pixel-channel combination).
